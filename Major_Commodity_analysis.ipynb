```python
import plotly.express as px
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
plt.style.use("seaborn-notebook")
```

```python
def close_price():
    fig = go.Figure()

    # Add Traces
    fig.add_trace(go.Scatter(x=Gold['Date']
                             , y=Gold['Close']
                             , name='Brent Oil - Close Pricing'
                             #, visible=True
                            ))

    fig.add_trace(go.Scatter(x=Palladium['Date']
                             , y=Palladium['Close']
                             , name='Palladium - Close Pricing'
                             #, visible=True
                            ))

    fig.add_trace(go.Scatter(x=Nickel['Date']
                             , y=Nickel['Close']
                             , name='Nickel - Close Pricing'
                             #, visible=True
                            ))

    fig.add_trace(go.Scatter(x=Brent_Oil['Date']
                             , y=Brent_Oil['Close']
                             , name='Brent Oil - Close Pricing'
                             #, visible=True
                            ))

    fig.add_trace(go.Scatter(x=Natural_Gas['Date']
                             , y=Natural_Gas['Close']
                             , name='Natural Gas - Close Pricing'
                             #, visible=True
                            ))

    fig.add_trace(go.Scatter(x=Wheat['Date']
                             , y=Wheat['Close']
                             , name='Wheat - Close Pricing'
                             #, visible=True
                            ))
    # Add Buttons
    fig.update_layout(
        updatemenus=[
            dict(
                active=0,
                buttons=list([ 

                    dict(label='Gold',
                         method='update',
                         args=[{'visible': [True, True, False, False, False, False, False, False]},
                               {'title': 'Close Pricing (Gold)'}]),

                    dict(label='Palladium',
                         method='update',
                         args=[{'visible': [False, False, True, True, False, False, False, False]},
                               {'title': 'Close Pricing  (Palladium)'}]),


                    dict(label='Nickel',
                         method='update',
                        args=[{'visible': [False,  False, False, False, True, True, False, False]},
                               {'title': 'Close Pricing  (Nickel)'}]),

                    dict(label='Brent Oil',
                         method='update',
                        args=[{'visible': [False,  False, False, False, True, True, False, False]},
                               {'title': 'Close Pricing  (Brent Oil)'}]),

                    dict(label='Natural Gas',
                         method='update',
                        args=[{'visible': [False,  False, False, False, True, True, False, False]},
                               {'title': 'Close Pricing  (Natural Gas)'}]),

                    dict(label='Wheat',
                         method='update',
                        args=[{'visible': [False,  False, False, False, True, True, False, False]},
                               {'title': 'Close Pricing  (Wheat)'}])
                ])
               )
        ])

    # Set slider selection buttons:
    fig.update_xaxes(
        rangeslider_visible=True,
        rangeselector=dict(
            buttons=list([
                dict(count=1, label='6m', step='month', stepmode='backward'),
                dict(count=6, label='1y', step='year', stepmode='backward'),
                dict(count=2, label='2y', step='year',  stepmode='backward'),
                dict(count=5, label='5y', step='year',  stepmode='backward'),
                dict(count=10, label='10y', step='year',  stepmode='backward'),
                dict(count=1, label='YTD',step='year', stepmode='todate')
            ])
        )
    )

    # Set title
    fig.update_layout(title_text='Close Pricing Average')

    # Set x and y-axis titles
    fig.update_xaxes(title_text='Date')
    fig.update_yaxes(title_text='Price ($)')

    fig.show()  
    
close_price()
```

```python
def CandleBar():
    # Candlestick charts for cryptocurrencies:
    fig = go.Figure()


    fig.add_trace(go.Candlestick(x=Palladium['Date']
                                 , open=Palladium['Open']
                                 , high=Palladium['High']
                                 , low=Palladium['Low']
                                 , close=Palladium['Close']
                                 , visible = False 
                                )) 

    fig.add_trace(go.Candlestick(x=Nickel['Date']
                                 , open=Nickel['Open']
                                 , high=Nickel['High']
                                 , low=Nickel['Low']
                                 , close=Nickel['Close']
                                 , visible = False 
                                ))

   fig.add_trace(go.Candlestick(x=Brent_Oil['Date']
                                 , open=Brent_Oil['Open']
                                 , high=Brent_Oil['High']
                                 , low=Brent_Oil['Low']
                                 , close=Brent_Oil['Close']
                                 , visible = False 
                                ))  
   fig.add_trace(go.Candlestick(x=Natural_Gas['Date']
                                 , open=Natural_Gas['Open']
                                 , high=Natural_Gas['High']
                                 , low=Natural_Gas['Low']
                                 , close=Natural_Gas['Close']
                                 , visible = False 
                                )) 
    fig.add_trace(go.Candlestick(x=Wheat['Date']
                                 , open=Wheat['Open']
                                 , high=Wheat['High']
                                 , low=Wheat['Low']
                                 , close=Wheat['Close']
                                 , visible = False 
                                ))


    # Add Buttons
    fig.update_layout(
        updatemenus=[
            dict(
                active=0,
                buttons=list([ 

                    dict(label='Gold',
                         method='update',
                         args=[{'visible': [True, False,False, False]},
                               {'title': 'Candlestick chart (Gold)'}]),

                    dict(label='Palladium',
                         method='update',
                         args=[{'visible': [False, True, False, False]},
                               {'title': 'Candlestick chart (Palladium)'}]),

                    dict(label='Brent_Oil',
                         method='update',
                         args=[{'visible': [False,  False, True, False]},
                               {'title': 'Candlestick chart (Brent Oil)'}]),

                    dict(label='Natural_Gas',
                         method='update',
                         args=[{'visible': [False, False, False, True]},
                               {'title': 'Candlestick chart (Natural Gas)'}]),
                    
                    dict(label='Wheat',
                         method='update',
                         args=[{'visible': [False, False, False, True]},
                               {'title': 'Candlestick chart (Wheat)'}]),
                    
                ]),
            )
        ])

    # Set slider selection buttons:
    fig.update_xaxes(
        rangeslider_visible=True,
        rangeselector=dict(
            buttons=list([
                dict(count=6, label='6m', step='month', stepmode='backward'),
                dict(count=1, label='1y', step='year', stepmode='backward'),
                dict(count=5, label='5y', step='year', stepmode='backward'),
                dict(count=1, label='YTD', step='year', stepmode='todate')
            ])
        )
    )


    # Set title
    fig.update_layout(title_text='Candlestick charts for selected cryptocurrency')

    # Set x and y-axis titles
    fig.update_xaxes(title_text='Date')
    fig.update_yaxes(title_text='Price ($)')

    fig.show()
    
CandleBar()
```
